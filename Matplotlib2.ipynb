```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('calories')
plt.ylabel('ounces')
plt.title('Calorie Weight Relationship')
```

```python
a =[1,1.5,2,2.5,3]
b =[5,7,10,5,0]

plt.plot(a,b)
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y, label='line1')


a =[1,1.5,2,2.5,3]
b =[5,7,10,5,0]
plt.plot(a,b, label='line2')

plt.legend()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y, label='line1',linestyle = 'dotted', linewidth=5, marker='x', markersize=12)


a =[1,1.5,2,2.5,3]
b =[5,7,10,5,0]
plt.plot(a,b, label='line2', linestyle = 'dashed', linewidth=4, marker='o', markersize=10)

plt.legend()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y, '*')

a =[1,1.5,2,2.5,3]
b =[5,7,10,5,0]

plt.plot(a,b, 'ro')
```
